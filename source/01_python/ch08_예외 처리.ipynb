# 1절. 예외처리 필요성
- 예외가 날 가능성이 있는 부분에 대해 미리 예상하고, 그에 대한 처리를 프로그래밍하는 것.
    (이유 : 안정적으로 실행할 수 있도록 함)
    
```
• 파일이 없거나 쓰기 금지로 인한 오류
• 데이터베이스 프로그래밍 시 제약조건 등에 의한 데이터베이스 서버 오류
• 네트워크 프로그래밍 시 네트워크 연결 실패로 인한 오류
• 리스트 또는 튜플의 인덱스를 벗어난 참조로 인한 오류
```

````python
filename = input('파일명?') # test.txt
# 한글 인코딩 방법 : 한글조합형(utf-8), 한글완성형(euc-kr < cp949) - euc-kr:2350개 문자 이외는 깨짐
# 한글문자 1만1172자(초성쌍자음 19개 * 중성 21 * 종성(ㄵ ㅀ같은거 포함) 28개 )
f = open('data/'+filename, 'r') # 읽기 전용으로 파일을 열기(읽기전용 stream객체 생성)
print(f.read())
f.close()
````

# 2절. try ~ except로 예외 처리
## 2.1 try ~ except
```
try:
    예외가 발생할 가능성이 있는 명령어들
except:
       예외가 발생했을 경우 실행할 명령어들
```

````python
# 100을 입력받은 정수값으로 나눠 출력하기
x = int(input('정수를 입력하세요 =>  '))
print('입력한 정수는 {} '.format(x))
print('100 을 입력한 정수로 나누면 {:.2f} '.format(100/x))
# ZeroDivisionError : "division by zero" 
#                      0을 입력했을때 예외 타입
# ValueError : invalid literal for int() with base 10: 'a'
#              invalid literal(잘못된 문자)를 입력할경우
````

````python
# 100을 입력받은 정수값으로 나눠 출력하기
# ZeroDivisionError
# ValueError
try:
    x = int(input('정수를 입력하세요 =>  '))
    print('입력한 정수는 {} '.format(x))
    print('100 을 입력한 정수로 나누면 {:.2f} '.format(100/x))
except:
    print('나누기에 유효한 정수가 아닙니다')
print('Done')
````

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100을 나눈 수를 출력
while True:
    try:
        x = int(input('정수를 입력하세요 =>  '))
        print('입력한 정수는 => {} '.format(x))
        print('100 을 입력한 정수로 나누면 => {:.2f} '.format(100/x))
        break
    except:
        print('나누기에 유효한 정수가 아닙니다')
````
