## 5.1 산술연산자:+-*/ //(몫연산자), **(제곱), %(나머지)

```python
print('10**5(승) =', 10**5)
print('4/2(결과 실수) =', 4/2)
print('4//2(몫)', 4//2)
print('4%2(나머지) =', 4%2)
```

```python
# 나누기를 제외한 산술연산은 정수끼리 결과는 정수. 정수와 실수 연산 결과는 실수
print('10**5.0 =', 10**5.0)
print('4*2 =', 4*2)
print('4*2.0 =', 4*2.0)
```

```python
# quiz1. 정수를 입력받아 짝수인지 홀수인지 여부를 출력
num = int(input('정수를 입력하세요 >'))
result = num%2
if result:
    temp = '홀수'
else:
    temp = '짝수'
print("입력하신 수는 {}. 2로 나눈 나머지는 {}, 그래서 {}를 입력하셨군요".format(num, num%2, temp))
print(f"입력하신 수는 {num}. 2로 나눈 나머지는 {num%2}, 그래서 {temp}를 입력하셨군요")
```

## 5.2 할당연산자 : =, +=, -= ,*=, /= , //=, %=

```python
a = 10
print(a)
a += 5 # a = a+5
print(a)
a -= 3 # a = a-3
print(a)
a *= 10 # a = a*10
print(a)
```

## 5.3 논리연산자 : and, &, or , |, not

```python
# | : 논리연산자, 비트연산자
print( (10>3) | (10>5) ) # 논리연산 True & True => True
print( 3 | 10 ) # 비트연산 수&수
#  3: 0 0 1 1
# 10: 1 0 1 0
# & : 1 0 1 1 (11)
print( 10>3 | 10<5) # 10>11>5 
```

```python
# and : 논리연산자, 거짓으로 판별되는 첫번때 항의 결과 반환
#       모든 항이 참이면 마지막항을 반환
print( True and True and 'abc')
print(0 and 5)
```

```python
# or : 논리연산자, 참으로 판별되는 첫번때 항의 결과 반환
#       모든 항이 거짓이면 마지막항을 반환
print( False or False or 0)
print(0 or '홍길동')
```

```python
not 10>5
```

# 6절 문자열 다루기
- 문자열 슬라싱변수.[from:stop:step]
- len(문자변수) : 문자수
- upper() : 대문자로
- lower() : 소문자로
- title() : 각 어절의 앞글자만 대문자로
- capitallize() : 첫문자만 대문자로
- count('찾을 문자열') : '찾을 문자열'이 몇개 들어있는지 갯수. 없으면 0
- find('찾을문자열') : '찾을문자열'이 왼쪽부터 시작하여 몇번째에 있는지 index. 없으면 -1
- rfind('찾을문자열') : '찾을문자열'이 오른쪽부터 시작하여 몇번째에 있는지 index. 없으면 -1
    cf.index('찾을문자열') : '찾을문자열'이 왼족부터 시작하여 몇번째에 있는지 index. 없으면 error
- startswith('특정문자열') : '특정문자열'로 시작하는지 여부
- endswith('특정문자열') : '특정문자열'로 끝나는지 여부

- strip() : 좌우 공백제거(trim)
- rstrip() : 오른쪽 공백제거
- lstrip() : 왼쪽 공백제거
- split(나눌sep) : 공백이나 다른 문자 기준으로 나누어 리스트로 반환
- isdigit() : 문자열이 숫자 문자열인지 여부
- islower() : 문자열이 전부 소문자인지 여부
- isupper() : 문자열이 전부 대문자인지 여부
- replace(oldSta, newStr) : 문자열안의 oldStr을 newStr로 변경한 값을 변환
