# 5절. 연산자
## 5.1 산술연산자:+-*/ //(몫연산자), **(제곱), %(나머지)

```python
print('10**5(승) =', 10**5)
print('4/2(결과 실수) =', 4/2)
print('4//2(몫)', 4//2)
print('4%2(나머지) =', 4%2)
```

```python
# 나누기를 제외한 산술연산은 정수끼리 결과는 정수. 정수와 실수 연산 결과는 실수
print('10**5.0 =', 10**5.0)
print('4*2 =', 4*2)
print('4*2.0 =', 4*2.0)
```

```python
# quiz1. 정수를 입력받아 짝수인지 홀수인지 여부를 출력
num = int(input('정수를 입력하세요 >'))
result = num%2
if result:
    temp = '홀수'
else:
    temp = '짝수'
print("입력하신 수는 {}. 2로 나눈 나머지는 {}, 그래서 {}를 입력하셨군요".format(num, num%2, temp))
print(f"입력하신 수는 {num}. 2로 나눈 나머지는 {num%2}, 그래서 {temp}를 입력하셨군요")
```

```python
# quiz2. 정수를 입력받아 5의 배수인지 여부를 출력
num = int(input('정수를 입력하세요 >'))
result = num%5
if result:
    temp = '배수X'
else:
    temp = '배수O'
print("입력하신 수는 {}. 5로 나눈 나머지는 {}, 그래서 {}".format(num, result, temp))
```

```python
# quiz2. 정수를 입력받아 5의 배수인지 여부를 출력
num = input('정수를 입력하세요=>')
if num [-1]=='0' or num[-1]=='5':
    print('5의 배수를 입력')
else:
    print('5의 배수 아님')
```

```python
kor = int(input('국어 점수는?=> '))
eng = int(input('영어 점수는?=> '))
mat = int(input('수학 점수는?=> '))
total = kor + eng + mat
avg = total / 3
```

## 5.2 할당연산자 : =, +=, -= ,*=, /= , //=, %=

```python
a = 10
print(a)
a += 5 # a = a+5
print(a)
a -= 3 # a = a-3
print(a)
a *= 10 # a = a*10
print(a)
```

## 5.3 논리연산자 : and, &, or , |, not

```python
# | : 논리연산자, 비트연산자
print( (10>3) | (10>5) ) # 논리연산 True & True => True
print( 3 | 10 ) # 비트연산 수&수
#  3: 0 0 1 1
# 10: 1 0 1 0
# & : 1 0 1 1 (11)
print( 10>3 | 10<5) # 10>11>5 
```

```python
# and : 논리연산자, 거짓으로 판별되는 첫번때 항의 결과 반환
#       모든 항이 참이면 마지막항을 반환
print( True and True and 'abc')
print(0 and 5)
```

```python
# or : 논리연산자, 참으로 판별되는 첫번때 항의 결과 반환
#       모든 항이 거짓이면 마지막항을 반환
print( False or False or 0)
print(0 or '홍길동')
```

```python
not 10>5
```

## 5.4 비교연산자 : >,>=, <, <=, ==, !=
- 연산자의 우선순위 고려(&, | -> 비교연산 -> and, or)

```python
address = "신림동"
age = 31
print(address=='신림동')
print((address!='신림동') & (age>=30))
```

```python
'진' > '변우석'
```

```python
'scott' < 'tiger'
```

```python
'a' < 'aa'
```
