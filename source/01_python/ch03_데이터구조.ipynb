```python
def save_index(lst, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst : 
        return lst.index(item)
    return -1
```

```python
def safe_remove(lst, remove_item):
    '''
    첫번째 매개변수 lst에서 remove_item 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst:
        lst.remove(remove_item)
    else:
        print('해당 데이터는 없는 데이터입니다')
```

## 3.2 딕셔너리 item 수정, 추가, 제거

```python
my_favorite = {'fruit':'apple', 'number':12, 'sports':'yoga'}
# 한 item씩 추가
my_favorite['people'] = '홍길동'
# my_favorite.get('people') 절대 사용X
print('1. 한 item 추가후 :',my_favorite)
# 여러 item을 한꺼번에 추가
my_favorite.update({'subject':'수학', 'major':'AI'})
print('2. 여러 item 추가후 :',my_favorite)

# 한 item씩 수정
my_favorite['fruit'] = 'banana'
print('3. 한 item 수정후 :',my_favorite)
# 여러 item을 한꺼번에 수정
# my_favorite.update({'fruit':'kiwi', 'number':15})
my_favorite.update(fruit='kiwi', number=15)
print('4. 여러 item 수정후 :', my_favorite)

# item 제거 방법1 : del 변수['삭제할key'] - 없는 key일 경우 에러
del my_favorite['fruit']
if 'fruit' in my_favorite:
    del my_favorite['fruit']
else:
    print('5. fruit 삭제하여 존재하지 않음')
# item 제거방법 2 : 변수.pop('삭제할key') - 해당 key값을 반환하고 제거 (없는 key일 경우 에러)
print('6. 제거한 데이터 :', my_favorite.pop('number'))
if 'number' in my_favorite:
    print('7. 제거한 데이터 :', my_favorite.pop('number'))
else:
    print('7. number 삭제하여 존재하지 않음')
print(my_favorite)
```

## 3.3 출력하기
- for문 이용

```python
my_tuple = ('홍길동', 30, '01099999999') # {0:'홍길동', 1:30, 2:'01099999999'}
for data in my_tuple:
    print(data, end='\t')
print() # 개행
for idx in range(len(my_tuple)): # [0,1,2]
    print("{}번째 : {}\t".format(idx, my_tuple[idx]), end="")
print() # 개행
for idx, value in enumerate(my_tuple): # (0,홍길동),(1,30),(2,'01099999999')
    print(f"{idx}번째 : {value}\t", end="")
```

## 4.4 셋연산자 : 합집합(|), 교집합(&), 차집합(-)

```python
names = ['길동', '길자', '순자']
tels = ['010-9999-9999', 
       '010-8888-8888',
       '010-7777-7777']
adds= ['서울', '인천', '판교']
for idx in range(len(names)): # [0,1,2]
    print("{}. 이름:{} / 연락처:{} / 지역 {}".format(idx+1,
                                           names[idx],
                                           tels[idx],
                                           adds[idx]))
```

```python
list(zip(names, tels,adds))
```

```python
for name, tel, add in zip(names, tels, adds):
    print(f"이름:{name} / 연락처: {tel} / 지역: {add}")
```

```python
list(enumerate(zip(names, tels, adds)))
```

```python
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    print(f"{idx+1}. 이름:{name} / 연락처:{tel} / 지역:{add}")
```

```python
# names, tels, adds를 딕셔너리 리스트로 
members = []
for name, tel, add in zip(names, tels, adds):
    member = {'name':name, 'tel':tel, 'address':add}
    members.append(member)
members
```

```python
members = []
for idx, (name, tel,add) in enumerate(zip(names, tels, adds)):
    member = {'순위':idx+1, '이름':name, '전화':tel, '주소':add}
    members.append(member)
members
```
