## 1.3 지역변수(local var)와 전역변수(global var)

````python
global_var = 100 # 전역변수
def func1():
    print('함수안 :', global_var)
func1()
print('함수밖 :', global_var)
````

````python
def func2():
    local_var = 200 # 지역변수
    print('함수안 :', local_var)
func2()
print('함수밖 :', local_var)
````

````python
# 변수의 값을 조회하는 순서 : 지역변수(지역심볼테이블) > 전역변수(전역심볼테이블) > 내장된 이름
g_var = 100 # 전역변수
def func3():
    g_var = 200 # 지역변수
    print('함수안 g_var :', g_var)
func3()
print('함수밖 g_var :', g_var)
````

````python
# 렉시컬 특징 : 변수가 어디에 선언(할당)되었는지에 따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    print('함수안 g_var :', g_var)
    g_var = 200 # 지역변수
func3()
print('함수밖 g_var :', g_var)
````

````python
# 렉시컬 특징 : 변수가 어디에 선언(할당)되었는지에 따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    global g_var # func3함수내의 g_var는 전역변수를 쓴다
    print('함수안 g_var :', g_var) # 전역변수
    g_var = 200 # 전역변수
func3()
print('함수밖 g_var :', g_var)
````

## 3.4 튜플 매개변수를 이용한 가변인수 설정

````python
# 함수 정의
def add(a, b):
    return a+b
def add(a, b, c):
    return a+b+c
````

````python
print(add(10, 20))
print(add(10, 2, 3))
````

````python
# args앞의 * : 매개변수글(인수들)이 args 튜플로 전달
def add(*args):
    "전달받은 매개변수들의 합을 return. 매개변수가 없을 경우 0을 return"
    #print(args) # (1, 2, 3)
    tot = 0 # 누적변수
    for num in args:
        tot += num # tot = tot + num
    return tot
````

````python
print(add())
print(add(1))
print(add(10, 20))
print(add(10, 2, 3))
print(add(1, 2, 3, 4, 5))
````

## 3.7 인수의 언패킹
### 튜플인수 언패킹
- 패킹 : 함수 정의시 매개변수 앞에 *붙임(튜플매개변수)
- 언패킹 : 함수 호출시 매개변수 앞에 *붙임(리스트나 튜플이 각자 요소별 매개변수로)

````python
def add(*args):
    tot = 0 
    for num in args:
        tot += num 
    return tot
````

````python
def map_template(func, L=[]):
    result = []
    for item in L:
        result.append(func(item))
    return result
````

````python
# "2000-01-01"을 받아 {'year':2000, 'month':1, 'day':1}을 return하는 함수
def split_date(date_): 
#     split_date = date_.split('-')
    return {'year':int(date_.split('-')[0]),
            'month':int(date_.split('-')[1]),
            'day':int(date_.split('-')[2])}
split_date = lambda date_ : {'year':int(date_.split('-')[0]),
                             'month':int(date_.split('-')[1]),
                             'day':int(date_.split('-')[2])}
````

````python
# {'name':'홍', 'birth':'2000-01-01'}을 받아 {'name':'홍', 'year':2000, 'month':1, 'day':1}을 return하는 함수
def split_birth(member):
    return {
        'name':member.get('name'),
        'year':int(member.get('birth').split('-')[0]),
        'month':int(member.get('birth').split('-')[1]),
        'day':int(member.get('birth').split('-')[2])
    }
split_birth({'name':'홍', 'birth':'2000-01-01'})
````

### (2) filter 

````python
def filter_template(func, L):
    result = []
    for item in L:
        if func(item):
            result.append(item)
    return result
````

````python
def even_chk(value): # 짝수면 True return, 홀수면 False를 return
    return value%2==0
even_chk = lambda value : value%2==0
even_chk(10)
````

````python
list_data = [1, 2, 3, 4, 5, 6, 7, 8]
filter_template(lambda value : value%2==0, list_data)
````

````python
[value for value in list_data if value%2==0]
````

````python
list(filter(lambda value : value%2==0, list_data))
````
