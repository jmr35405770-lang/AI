정규표현식(regex)
```(마크다운에서 백틱은블록을만들때사용)
    1. 참조 : https://ko.wikipedia.org/wiki/정규_표현식 위키백과
    2. 간략문법
    \d : (숫자와매치. [0-9]와 동일)
    \D : (숫자가아닌것)
    \w : (영문자나숫자[a-zA-Z0-9]와 동일)
    \W : (영문자나숫자가아닌문자) 
    . : (문자)
    \. : (.)
    \s : ( )
    \- : (-)
    {2,} : (2번 이상 반복)
    {2, 4} : (2~4회 반복) #(4,4)는 불가능
    ? : (0번이나 1번 의미)
    + : (1번이상반복)
    * : (0번이상반복)
    3. 정규표현식연습장: https://regexr.com/
    ex1. 전화번호 : 02 888 7777 / 028887777 / 010-9999-9999 / 010.9999.9999 / 
            02)999-9999
            => 0\d{1,2}[ \-\.)]?\d{3,4}[ \-\.]?\d{4}
    ex2. 이메일 : abc12@naver.com / abc12@sns.co.kr
            => \w+@\w+(\.\w+){1,2}
    ex3. 주민번호 : 981212-1052525
        => \d{2}(0[1-9]|1[0-2])(0[1-9]|[12]\d|30|31)-[1-4]\d{6}
    4. 소스 구현
```

````python
data = "이름:홍길동, 주소:신림동, 전화:02-888-8888, \
핸드폰:010-9999-9999, 이메일:abc@lg.co.kr, 이메일:abc@naver.com\
특징:동해번쩍 서해번쩍, 혈액형:B ㅋㅋ"
````

````python
import re
tel_pattern = r"0\d{1,2}[\-)\s\.]\d{3,4}[\-\s\.]\d{4}"
tels = re.findall(tel_pattern, data) # 해당 패턴을 다 찾아줘. 없으면 빈 list([])
tels
````

````python
# 정규표현식 패턴과 처음 match되는 것을 찾아줘
firstTel = re.search(tel_pattern, data)
firstTel
````

````python
print(firstTel.group(0))
print(data[19:30])
````

````python
phone_pttern = r"(\d{2,3})-(\d{3,4})-(\d{4})"
phone = re.search(phone_pttern, data)
print(phone.group(0))
print(phone.group(1))
print(phone.group(2))
print(phone.group(3))
````

````python
email_pttern = r"\w[a-z0-9_]+@\w+(\.\w+){1,2}"
print(re.findall(email_pttern, data))
print(re.search(email_pttern, data).group(0))
````

````python
# 특정 문자외 없애기
re.sub(r"[^가-힣0-9\-a-z@]"," ", data)
````

## 4.2 format()문법

- "{[인덱스]:[공백대체문자][정렬방식][자릿수][타입]}".format(변수명)
    * 인덱스 : format 함수 안의 매개변수 순수
    * 공백대체문자는 정렬방식과 자릿수를 명시하여야 함
    * 정렬방식 : <(왼쪽정렬), >(오른쪽정렬), ^(가운데정렬)
    * 정수자릿수 : 출력할 자리수
    * 실수자릿수 : 전체출력될자리수.변수중출력할자리수
    * 타입 : d(10진수), f(실수4.1f:1은 소수점자리수), s(문자), o(8진수), x(16진수), b(2진수)

````python
name = input("이름은?")
age = int(input("나이?"))
````

````python
print("[{0:^10}]살 [{1:^10}]님 안녕하세요".format(age,name))
print(f"[{age:^10}]살 [{name:^10}]님 안녕하세요")
````

````python
a = 2.755
sign = '+'
width = 10
digits = 3
print("[{:+10.3}]".format(a))
print("[{:{sign}{width}.{digits}}]".format(a, 
                                           sign=sign,
                                           width=width,
                                           digits=digits))
````

## 4.3 날짜출력

````python
from datetime import datetime
i = 10
day = datetime(2026, 11, 19, 18, 0, 0) # 날짜형데이터에는 날짜와 시간이 다 저장
print(day)
````

````python
# Y(2026) y(26) m(월) d(일) H(24시) M(분) S(초) %I(12시) %p(am/pm)
print("{:%Y년%m월%d일%H시%M분%S초}".format(day))
print("{:%Y년%m월%d일 %p %I시%M분%S초}".format(day))
print(day.strftime("%Y년%m월%d일 %p %I시%M분%S초"))
# 날자를 문자열로 전환하는 함수 : 날짜형변수.strftime(포맷)
````
