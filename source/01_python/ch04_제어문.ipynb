# 1절. 조건문(if)
```
if 조건1:
    조건1이 참일 경우 수행할 명령어들(if문 단독 가능. 들여쓰기 중요)
elif조건2:
    조건1이 거짓이고, 조건2가 참일경우 수행할 명령어들(elif는 필요한만큼 사용가능)
else:
    조건이 만족하지 않으 때 수행할 명령어들(else는 단독사용불가)
```

## 2.1 for-in문
```
for 임시변수 in 나열가능한자료 :
    반복할 문장들
else:
    나열 가능한 자료를 다 소요하여 for블록을 다 수행한 후, 최종 한번 실행할 문장들
    for문안의 break를 수행할 경우 , else블록을 수행하지 않음(else는 선택)
```

````python
for i in range(1, 10, 2): # [1, 3, 4, 7, 9]
    print(i, end="\t")
else:
    print("\nDONE")
    print('END')
````

````python
for i in range(1, 10):
    if i==5:
        break # 현재 반복문을 빠져나오는 명령
    print(i, end="\t")
else:
    print('\nDONE')
print('END')
````

## 2.2 while문
```
while 조건:
    조건이 참일 경우 반복할 문장들
else: 
    조건이 거짓일 경우 한번 실행할 문장(else 선택사항)
```

````python
# 1~100까지 짝수들의 합을 계산하고 출력하시오(while문)
tot = 0 # 누적변수
i = 1
while i<=100:
    if i%2==0:
        tot += i
    i += 1 # i변수 1증가
else:
    print('1~100까지 짝수의 합은', tot)
````

````python
# - 사용자에게두정수를입력받아작은수부터큰수까지홀수들의합을누적하여출력하시오
first = int(input('첫번째 정수는 : '))
second = int(input('두번째 정수는 : '))
# (1) 두 수 중 작은 수는 minNo, 큰수는 maxNo에 할당
if first<second:
    minNo = first
    maxNo = second
else:
    minNo = second
    maxNo = first
# minNo = first if first<second else second
# maxNo = first if first>second else second
# (2) minNo~maxNo 까지 홀수 누적하기
tot = 0
for no in range(minNo, maxNo+1):
    if no%2: # if no%2!=0
        tot += no
print(f"{minNo}부터 {maxNo}까지 홀수의 합은 {tot}입니다")

````

````python
# - 사용자에게두정수를입력받아작은수부터큰수까지홀수들의합을누적하여출력하시오
first = int(input('첫번째 정수는 : '))
second = int(input('두번째 정수는 : '))
# (1) first에 작은수를 second에 큰수
if first>second:
    # 바꿔
#    temp = first
#    first = second 
#    second = temp 
    first, second = second, first
# (2) first부터 second까지 홀수합 누적
tot = 0
for no in range(first, second):
    if no%2: # if no%2!=0
        tot += no
print(f"{first}부터 {second}까지 홀수의 합은 {tot}입니다")
````

````python
import random
# random.randint(1, 45)를 최소 6개 이상 => 리스트,셋 => 정렬괸 값 출력
lotto_lst = []
cnt = 0 # 난수 발생 횟수
while len(lotto_lst)<6:
    cnt += 1
    temp = random.randint(1, 45) # 난수
    if temp not in lotto_lst: # temp가 lotto_lst의 요소가 아닌지 여부
        lotto_lst.append(temp)
print('정렬된 로또번호 :',sorted(lotto_lst))
print('while문 수행 횟수(난수 발생 횟수) :', cnt)
````

````python
# 1~100사이의 정수를맞추는프로그램을구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메시지 뿌리고 끝남(1~100 사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하 메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메세지 뿌리고 끝남
'''
import random
randomNum = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 :', randomNum)
while True:
    inputNum = int(input("1~100사이의 수를 맞춰보세요 :"))
    if (inputNum>100) | (inputNum<1):
        print('수의 범위를 벗어났습니다')
    elif inputNum>randomNum:
        print("DOWN ↓ ")
    elif inputNum<randomNum:
        print("UP ↑ ")
    else:
        print('축하 축하. 맞추셨습니다')
        break
print('DONE')
````

````python
# 1~100사이의 정수를맞추는프로그램을구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메시지 뿌리고 끝남(1~100 사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하 메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메세지 뿌리고 끝남
'''
import random
randomNum = random.randint(1, 100)
# print('컴퓨터가 발생한 임의의 수 :', randomNum)
inputNum = 0
while inputNum!=randomNum:
    inputNum = int(input("1~100사이의 수를 맞춰보세요 :")) # 2단계는 min~max사이의 수를 맞춰보세요
    if (inputNum>100) | (inputNum<1):
        print('수의 범위를 벗어났습니다')
    elif inputNum>randomNum:
        print("DOWN ↓ ")
    elif inputNum<randomNum:
        print("UP ↑ ")
    else:
        print('축하 축하. 맞추셨습니다')
print('DONE')
````

# 4절. 중첩루프 탈출

````python
for a in range(0,3): # a <= 0, 1, 2
    for b in range(1, 3): # b <=1 ,2
        if a == b:
            print("a={}, b={} break".format(a, b))
            break
        print(a, b)

print("DOWN")        
````

````python
# 중첩 루프 탈출 방법1 : flag변수 이용
flag = False
for a in range(0,3): 
    for b in range(1, 3): 
        if a == b:
            print("a={}, b={} break".format(a, b))
            flag = True
            break
        print(a, b)
    if flag:
        break
        
print("DOWN")  
````

````python
# 중첩 루프 탈출 방법2 : Exception(예외처리) 이용(7/7일 예정)
class BreakOutLoopError(Exception):
    pass
try:
    for a in range(0, 3): # a = 0, 1, 2
        for b in range(1, 3): # b = 1, 2
            if a==b:
                raise BreakOutLoopError() # 강제예외발생
            print(a, b)
except BreakOutLoopError:
    pass
print('Done')
````
